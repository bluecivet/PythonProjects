# initaling data

```python
import pandas as pd
import numpy as np

testTable = pd.read_csv("test.csv", engine = "python", encoding = "big5", names = [1,2,3,4,5,6,7,8,9])

testTable.head(18)
```

```python
weightTable = pd.read_csv("weights.csv")
weightTable
```

```python
biasTable = pd.read_csv("bias.csv")
biasTable
```

```python
w = np.array(weightTable.iloc[0,1:])
w
```

```python
b = np.array(biasTable.iloc[0,1:])
b
```

```python
testTable.replace("NR", "0", inplace = True)
```

```python
def readData(df, rowStep, columnStep):
    rowNum = df.shape[0]
    columnNum = df.shape[1]
    output = []
    for i in range(0, rowNum, rowStep):
        dataTable = df.iloc[i:i + rowStep, 0:columnStep]
        arr = np.array(dataTable, dtype = "float")
        arrAvg = arr.mean(axis = 1)
        output.append(arrAvg.reshape(18,1))
    return output
```

```python
xInput = readData(testTable, 18, 9)
xInput
```

# start getting result

```python
label = []
pm = []  # pm 2.5

for i in range(len(xInput)):
    label.append("id" + str(i))
    xArr = xInput[i]
    y = np.dot(w, xArr) + b 
    pm.append(y)

print(label)
print(pm)
```

```python
dict = {'id':label, 'pm2.5' : pm}
outputTable = pd.DataFrame(dict)
outputTable
```

```python
outputTable.to_csv("output.csv")
```
